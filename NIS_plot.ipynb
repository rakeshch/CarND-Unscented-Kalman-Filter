# Import required libraries

```python
import matplotlib.pyplot as plt
```

# Read Lidar and radar NIS files

```python
nis_rad_list = []
nis_lid_list = []

with open("build/NIS_radar.txt") as file:
    for line in file: 
        line = line.strip()
        if(len(line) > 0):
            nis_lid_list.append(float(line))
            
with open("build/NIS_lidar.txt") as file:
    for line in file: 
        line = line.strip()        
        if(len(line) > 0):
            nis_rad_list.append(float(line))
```

# Plot NIS vs Time Step

```python
plt.figure(figsize=(20,5))
plt.subplot(1, 2, 1)
plt.plot(nis_rad_list)
plt.ylabel('NIS Value')
plt.xlabel('Time Step')
plt.title('Radar NIS Values')
plt.axhline(y=7.8, color='r', linestyle='-')
plt.ylim([0,14])

plt.subplot(1, 2, 2)
plt.plot(nis_lid_list)
plt.ylabel('NIS Value')
plt.xlabel('Time Step')
plt.title('Lidar NIS Values')
plt.axhline(y=7.8, color='r', linestyle='-')
plt.ylim([0,14])

plt.show()
```
